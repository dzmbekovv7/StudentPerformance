```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import joblib
from config import DATASET_PATH, PASS_FAIL_MODEL_PATH
```

```python
df = pd.read_csv(DATASET_PATH)
```

```python
df.head()
```

```python
X = df[["Study_Hours_Per_Day","Attendance_Percentage","Math_Score","Science_Score","English_Score","Previous_Year_Score"]]
y = df["Pass_Fail"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3
)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
joblib.dump(model, PASS_FAIL_MODEL_PATH)
```
